```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.zip")
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df[['Survived','Pclass','Fare','Sex','Age']].isnull().sum())
```

```python
print(df.columns)
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
print(df[['Age']].head())
```

```python
df['Sex']=df['Sex'].map({'Male':1,'female':0})
```

```python
print(df[['Sex']].head())
```

```python
df['Sex']=df['Sex'].fillna(df['Sex'].mean())
```

```python
print(df[['Sex']].head())
```

```python
X=df[['Pclass','Sex','Age','Fare']]
Y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_test,X_train,Y_test,Y_train=train_test_split(X,Y,test_size=0.2,random_state=30)
```

```python
print(X_test.head())
```

```python
print(X_train.head())
```

```python
print(Y_test.head())
```

```python
print(Y_train.head())
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(max_iter=100,random_state=30)
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(Y_test,Y_pred)
```

```python
print("accuracy = ",accuracy)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(Y_test,Y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
```

```python
classification=classification_report(Y_test,Y_pred)
```

```python
print(classification)
```
